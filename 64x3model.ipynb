```
import tensorflow as tf
from tensorflow.keras.callbacks import TensorBoard
from tensorflow.keras.models import Sequential
import numpy as np
from tensorflow.keras.layers import Dense,Dropout,Activation,Flatten,Conv2D,MaxPooling2D
import pickle
import time
```

```
x = pickle.load(open("/content/drive/My Drive/ASl detector/dataset/x3.pickle",'rb'))
y = pickle.load(open("/content/drive/My Drive/ASl detector/dataset/y3.pickle",'rb'))
```

```
#normalise the data 
#our data is image data is min is 0 and max is 255 

x=x/255.0

#convert the y into an array otherwise it will show an error
y=np.array(y)
```

```
tensorboard = TensorBoard(log_dir='/content/drive/My Drive/ASl detector/logs/64x3model')
```

```
model = Sequential()
```

```
#layer 1
model.add(Conv2D(128 , (5,5) , input_shape = x.shape[1:] , activation = 'relu' ))
model.add(MaxPooling2D(pool_size = (2,2)))

#layer 2
model.add(Conv2D(64,(3,3), activation = 'relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Dropout(0.1))

#layer 3
model.add(Flatten())

#layer 4
model.add(Dense(64 , activation='relu'))
model.add(Dropout(0.1))

#layer 5
model.add(Dense(32 , activation='relu'))

#output layer 
model.add(Dense(29  , activation='softmax'))
```

```
model.compile(loss='sparse_categorical_crossentropy' , optimizer='adam' , metrics=['accuracy'])
```

```
model.fit(x,y, batch_size=16 , epochs=30 , validation_split=0.3 , callbacks=[tensorboard])
```

```
model.save('64x3-CNN_m1.model')
```
