```python
categories = ['A','B','C','D','del','E','F','G','H','I','J','K','L','M','N',
              'nothing','O','P','Q','R','S','space','T','U','V','W','X','Y','Z']
```

```python
def preproc(img):
    minValue=70
    blur = cv2.GaussianBlur(img, (5, 5), 2)
    th3 = cv2.adaptiveThreshold(blur, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 13, 2)
    ret, res = cv2.threshold(th3, minValue, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
    return res



def prepare(filepath):
    IMG_SIZE = 50
    img_arr = cv2.imread(filepath,cv2.IMREAD_GRAYSCALE)
    new_array = cv2.resize(img_arr,(IMG_SIZE,IMG_SIZE))
    new_array = preproc(new_array)
    return new_array.reshape(-1,IMG_SIZE,IMG_SIZE,1)
```

```python
model = tf.keras.models.load_model(r"D:\STUDY 3\soft computing\ASL project\models\64x3-CNN_m1.model")
```

```python
for image in os.listdir(path):
    input_path = os.path.join(path,image)
    print(input_path[81:])
    img = cv2.imread(input_path)
    predict = model.predict([prepare(input_path)])
    plt.figure(figsize=(10,8))
    plt.imshow(predict, cmap = plt.cm.binary)
    plt.xticks(np.arange(len(categories)) , categories)
    plt.show()
    cv2.imshow('df',img)
    cv2.waitKey(0)

cv2.destroyAllWindows()
```
